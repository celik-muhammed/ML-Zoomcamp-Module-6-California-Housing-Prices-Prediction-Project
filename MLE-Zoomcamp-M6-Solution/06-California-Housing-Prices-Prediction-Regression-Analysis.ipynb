**About Dataset**

Median house prices for California districts derived from the 1990 census.

**Context**

This is the dataset used in the second chapter of Aurélien Géron's recent book 'Hands-On Machine learning with Scikit-Learn and TensorFlow'. It serves as an excellent introduction to implementing machine learning algorithms because it requires rudimentary data cleaning, has an easily understandable list of variables and sits at an optimal size between being to toyish and too cumbersome.

The data contains information from the 1990 California census. So although it may not help you with predicting current housing prices like the Zillow Zestimate dataset, it does provide an accessible introductory dataset for teaching people about the basics of machine learning.

**Content**

The data pertains to the houses found in a given California district and some summary stats about them based on the 1990 census data. Be warned the data aren't cleaned so there are some preprocessing steps required! The columns are as follows, their names are pretty self explanitory:

longitude

latitude

housing_median_age

total_rooms

total_bedrooms

population

households

median_income

median_house_value

ocean_proximity

### DictVectorizer

````python
# from sklearn.preprocessing import OneHotEncoder
# from sklearn.compose import make_column_transformer
from sklearn.preprocessing import FunctionTransformer
from sklearn.feature_extraction import DictVectorizer

from sklearn.tree import DecisionTreeRegressor
from sklearn.pipeline import Pipeline

# Define a function to transform DataFrame to dictionary records
def to_dict_records(df: pd.DataFrame):
    return df.to_dict(orient='records')

def model_builder():
    # Create a preprocessor to handle categorical columns with one-hot encoding
    preprocessor = Pipeline([
        ('to_dict', FunctionTransformer(func=to_dict_records)),
        ('dv', DictVectorizer(sparse=True)),
    ])
    # Create a Decision Tree Regression model
    model = Pipeline([
        ('preprocessor', preprocessor),
        ('dt', DecisionTreeRegressor(max_depth=1))
    ])
    return model

# Create a model
model = model_builder()

# Fit the model on the training dataset
model.fit(X_train, y_train)

# Get last step feature dimensionality, Display the dimensionality
print("number of columns:", len(model[0][-1].get_feature_names_out()))
````

## Question 2

Train a random forest model with these parameters:

* `n_estimators=10`
* `random_state=1`
* `n_jobs=-1` (optional - to make training faster)


What's the RMSE of this model on validation?

* 0.045
* 0.245
* 0.545
* 0.845

````python
# from sklearn.preprocessing import OneHotEncoder
# from sklearn.compose import make_column_transformer
from sklearn.preprocessing import FunctionTransformer
from sklearn.feature_extraction import DictVectorizer

# from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline

# Define a function to transform DataFrame to dictionary records
def to_dict_records(df: pd.DataFrame):
    return df.to_dict(orient='records')

def model_builder(params=dict(n_estimators=10)):
    # Create a preprocessor to handle categorical columns with one-hot encoding
    preprocessor = Pipeline([
        ('to_dict', FunctionTransformer(func=to_dict_records)),
        ('dv', DictVectorizer(sparse=True)),
    ])
    # Create a random forest regressor with the specified parameters
    model = Pipeline([
        ('preprocessor', preprocessor),
        ('rf',RandomForestRegressor(random_state=1, n_jobs=-1, **params))
    ])
    return model

# Create a model
model = model_builder()

# Fit the model on the training dataset
model.fit(X_train, y_train)

# Get last step feature dimensionality, Display the dimensionality
print("number of columns:", len(model[0][-1].get_feature_names_out()))
````

````python
# Make predictions on the validation set
y_val_pred = model.predict(X_val)

# Calculate RMSE
rmse = mean_squared_error(y_val, y_val_pred, squared=False)

# Display the RMSE
print("RMSE on validation:", rmse)
````

## Question 3

Now let's experiment with the `n_estimators` parameter

* Try different values of this parameter from 10 to 200 with step 10.
* Set `random_state` to `1`.
* Evaluate the model on the validation dataset.


After which value of `n_estimators` does RMSE stop improving?
Consider 3 decimal places for calculating the answer.

- 10
- 25
- 50
- 160

````python
from tqdm.auto import tqdm

# Define a range of n_estimators values to test
# Initialize variables to keep track of the best RMSE and the corresponding n_estimators
n_estimators_values = range(10, 201, 10)
best_n_estimators = None
best_rmse = float("inf")
scores = []

for n_estimators in tqdm(n_estimators_values):
    # Create a random forest regressor with the specified parameters
    model = model_builder({'n_estimators': n_estimators})
    # Train the model on the training data
    model.fit(X_train, y_train)

    # Make predictions on the validation set
    y_val_pred = model.predict(X_val)

    # Calculate RMSE
    score = mean_squared_error(y_val, y_val_pred, squared=False)
    scores.append((n_estimators, score))

    # Check if this RMSE is better than the current best
    if score < best_rmse:
        best_rmse = score
        best_n_estimators = n_estimators

df_scores = pd.DataFrame(scores, columns=['n_estimators', 'rmse'])
# Display the n_estimators after which RMSE stops improving
print(f"After which value of n_estimators does RMSE ({best_rmse}) stop improving: {best_n_estimators}")
````

````python
plt.figure(figsize=(12, 5))
plt.plot(df_scores.n_estimators, df_scores.rmse, label='rmse', color='C0')
plt.plot(df_scores.n_estimators, df_scores.rmse.round(3), label='rmse_rounded_3', color='C1')
plt.xticks(df_scores.n_estimators.unique())
plt.xlabel('n_estimators')
plt.ylabel('rmse')
plt.legend()
plt.show()
````

### We can use `warm_start=True` to train faster. 

> In this case, we train only additional `10 new trees` each time, not the entire ensemble from scratch like previously

```py
warm_start : bool, default=False
    When set to ``True``, reuse the solution of the previous call to fit
    and add more estimators to the ensemble, otherwise, just fit a whole
    new forest. See :term:`the Glossary <warm_start>`.
```

````python
from tqdm.auto import tqdm

# Define a range of n_estimators values to test
# Initialize variables to keep track of the best RMSE and the corresponding n_estimators
n_estimators_values = range(10, 201, 10)
best_n_estimators = None
best_rmse = float("inf")
scores = []

# Create a random forest regressor with the specified parameters
model = model_builder({'n_estimators': 1, 'warm_start': True})

for n_estimators in tqdm(n_estimators_values):
    model[-1].n_estimators = n_estimators
    # Train the model on the training data    
    model.fit(X_train, y_train)

    # Make predictions on the validation set
    y_val_pred = model.predict(X_val)

    # Calculate RMSE
    score = mean_squared_error(y_val, y_val_pred, squared=False)
    scores.append((n_estimators, score))

    # Check if this RMSE is better than the current best
    if score < best_rmse:
        best_rmse = score
        best_n_estimators = n_estimators

df_scores = pd.DataFrame(scores, columns=['n_estimators', 'rmse'])
# Display the n_estimators after which RMSE stops improving
print(f"After which value of n_estimators does RMSE ({best_rmse}) stop improving: {best_n_estimators}")
````

````python
plt.figure(figsize=(12, 5))
plt.plot(df_scores.n_estimators, df_scores.rmse, label='rmse', color='C0')
plt.plot(df_scores.n_estimators, df_scores.rmse.round(3), label='rmse_rounded_3', color='C1')
plt.xticks(df_scores.n_estimators.unique())
plt.xlabel('n_estimators')
plt.ylabel('rmse')
plt.legend()
plt.show()
````

````python
%%time

from tqdm.auto import tqdm
# Define the values of max_depth and n_estimators to test
# Initialize variables to keep track of the best max_depth and the corresponding mean RMSE
max_depth_values = [10, 15, 20, 25]
n_estimators_values = range(10, 201, 10)
best_max_depth = None
best_mean_rmse = float("inf")
scores = []


for max_depth in tqdm(max_depth_values):
    mean_rmse = 0
    # Create a random forest regressor with the specified parameters
    model = model_builder({'n_estimators': 0, 'max_depth': max_depth, 'warm_start': True})

    for n_estimators in tqdm(n_estimators_values):
        model[-1].n_estimators = n_estimators
        # Train the model on the training data
        model.fit(X_train, y_train)

        # Make predictions on the validation set
        y_val_pred = model.predict(X_val)

        # Calculate RMSE
        score = np.sqrt(mean_squared_error(y_val, y_val_pred))
        scores.append((max_depth, n_estimators, score))

    # Calculate the mean RMSE for this max_depth
    mean_score = np.array(scores)[:, 2].mean()
    # Check if this mean RMSE is better than the current best
    if mean_score < best_mean_rmse:
        best_mean_rmse = mean_score
        best_max_depth = max_depth


columns = ['max_depth', 'n_estimators', 'rmse']
df_scores = pd.DataFrame(scores, columns=columns)
# Display the best max_depth based on the mean RMSE
print(f"Best max_depth using mean RMSE: {best_max_depth}")
````

Which eta leads to the best RMSE score on the validation dataset?

````python
# from sklearn.preprocessing import OneHotEncoder
# from sklearn.compose import make_column_transformer
from sklearn.preprocessing import FunctionTransformer
from sklearn.feature_extraction import DictVectorizer

import xgboost as xgb
from sklearn.pipeline import Pipeline

def xgb_train_parameters(train_df, y_train, val_df, y_val, params={}):
    # Define a function to transform DataFrame to dictionary records
    def to_dict_records(df: pd.DataFrame):
        return df.to_dict(orient='records')
    
    # Create a preprocessor to handle categorical columns with one-hot encoding
    preprocessor = Pipeline([
        ('to_dict', FunctionTransformer(func=to_dict_records)),
        ('dv', DictVectorizer(sparse=True)),
    ]).fit(train_df, y_train)

    # Create a DMatrix for XGBoost
    features = [f.replace('<', 'less ') for f in preprocessor[-1].feature_names_]
    dtrain   = xgb.DMatrix(data=preprocessor.fit_transform(train_df), label=y_train, feature_names=features)
    dval     = xgb.DMatrix(data=preprocessor.transform(val_df), label=y_val, feature_names=features)
    
    # Create a watchlist for validation
    watchlist = [(dtrain, 'train'), (dval, 'val')]

    current_params = {
        'eta': 0.1,
        'max_depth': 6, 'min_child_weight': 1, 
        'objective': 'reg:squarederror', 
        'nthread': 8, 'seed': 1, 'verbosity': 1,
    }
    # Update params
    current_params.update(params)
    return dict(params=current_params, dtrain=dtrain, evals=watchlist, num_boost_round=100, early_stopping_rounds=25, verbose_eval=5)

# Get booster parameters
booster_params = xgb_train_parameters(X_train, y_train, X_val, y_val)
dtrain = booster_params['evals'][0][0]
dval = booster_params['evals'][1][0]

model = xgb.train(**booster_params)
````

````python
def parse_xgb_output(output):
    results = []

    for line in output.stdout.strip().split('\n'):
        it_line, train_line, val_line = line.split('\t')

        it = int(it_line.strip('[]'))
        train = float(train_line.split(':')[1])
        val = float(val_line.split(':')[1])

        results.append((it, train, val))
    
    columns = ['num_iter', 'train_auc', 'val_auc']
    df_results = pd.DataFrame(results, columns=columns)
    return df_results
````
